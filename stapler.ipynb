```python
# Load packages and set pi
import math as m
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.lines import Line2D
pi = m.pi
```

```python
min_x = []
min_y = []
min_d = []
min_a = []
y_values = []
x_values = []

for y in range(1,296):
    print(round(y/296,2))
    for x in range(1,209):
        angle_values = []
        area_values = []
        total_x_values = []
        total_y_values = []
        for angle in np.arange(0.0001, (pi/2)-0.0001, 0.0001):
            y_2 = x/m.tan(angle)
            x_2 = y*m.tan(angle)
            total_x = x_2 + x
            total_y = y_2 + y
            
            # Make sure dimensions fit in an A4 paper 
            
            if total_y<297 and total_x<210:
                area = (total_x*total_y)/2
                area_values.append(area)
                angle_values.append(m.degrees(angle))
                total_x_values.append(total_x)
                total_y_values.append(total_y)
                
            else:
                continue
                
        # Make sure there IS a minimum area to add
        
        if len(area_values)>0:
            min_x.append(total_x_values[area_values.index(min(area_values))])
            min_y.append(total_y_values[area_values.index(min(area_values))])
            min_d.append(angle_values[area_values.index(min(area_values))])
            min_a.append(min(area_values))
            y_values.append(y)
            x_values.append(x)
```

```python
# Background Color
ax = plt.axes()
ax.set_facecolor("#EBE5C1")

# Create the Plot and Grid
plt.scatter(min_d, min_a, color='#031D3A', label='Area')
plt.grid(linestyle='--', alpha = 0.7,c="#E3B441")
plt.ylim(0, max(min_a)+(max(min_a)*0.1))

# Title
plt.suptitle('Distribution of Minimum Areas over Angles', x=0.125, y=0.98, ha='left', fontsize=18)

# Subtitle
plt.title('Points of which a local minimum exists', loc='left', fontsize=14)

# Bottom Text
text = 'The graph above shows the minimum area possible for each possible position of the staple. The angle shows how to attain the minimum area.'
plt.text(0, -6000, text, ha='left', fontsize = 11, 
         alpha=0.9)

# Axis Labels
plt.ylabel('Total Area (mm)',fontsize=15)
plt.xlabel('Angles (degrees)',fontsize=15)

# Annotation Arrow
arrowprops = dict(arrowstyle="->", connectionstyle="angle3,angleA=0,angleB=-90")
plt.annotate('Biggest "Minimum" Area', 
             xy=(min_d[min_a.index(max(min_a))],max(min_a)), 
             xytext=(min_d[min_a.index(max(min_a))]+5, max(min_a)+5),
             arrowprops=arrowprops,
             size = 20)

# Annotation Line
line_y = min_d[min_a.index(max(min_a))]
plt.axvline(line_y, linestyle='--', color='#E3B441', alpha=0.8,linewidth = 2.5)

# Size and Show
plt.rcParams['figure.figsize'] = [18, 6]
plt.xticks(range(0, 90, 5))
plt.savefig('aa.png', dpi=300, bbox_inches='tight')
```

```python
ax = plt.axes()
ax.set_facecolor("#EBE5C1")
hist, bins, patches = plt.hist(min_d, bins=90, edgecolor='#7F9384', color='#074388')

# Find the bin with the highest frequency and draw the line
max_bin_idx = np.argmax(hist)
highest_bin_value = bins[max_bin_idx]
highest_bin_frequency = hist[max_bin_idx]

# Draw vertical line at the highest point
plt.axvline(x=highest_bin_value, color='#E3B441', linestyle='--', alpha=0.8,linewidth = 2.5)

# arrow
arrowprops = dict(arrowstyle="->", connectionstyle="angle3,angleA=0,angleB=-90", mutation_scale=20)
plt.annotate('Most Common Degree', 
             xy=(highest_bin_value,highest_bin_frequency+1), 
             xytext=(highest_bin_value+5, highest_bin_frequency),
             arrowprops=arrowprops,
             size = 20)

plt.xlabel('Degree Values')
plt.ylabel('Frequency')
plt.title('Histogram of Degree Values')
plt.xticks(range(0, 90, 5))

plt.ylim(0, highest_bin_frequency+(highest_bin_frequency*0.1))

plt.savefig('bb.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
## loop

color_1='#FF5555'
color_2='#C44E65'
color_3='#894875'
color_4='#4E4184'
color_5='#133A94'

# Create the figure
x_length = 210
y_length = 297
fig, ax = plt.subplots(figsize=(x_length/35, y_length/35))
ax.set_aspect('equal')
ax.set_facecolor("#EBE5C1")

for i in range(0, len(x_values)):
    # Coordinates of the point
    x_point = x_values[i]
    y_point = y_values[i]

    x_intercept = min_x[i]
    y_intercept = min_y[i]

    # Plot the point
    if min_d[i] <18:
        ax.plot(x_point, y_point, marker='o', color=color_1)
    elif min_d[i] <36:
        ax.plot(x_point, y_point, marker='o', color=color_2)
    elif min_d[i] <54:
        ax.plot(x_point, y_point, marker='o', color=color_3)
    elif min_d[i] <72:
        ax.plot(x_point, y_point, marker='o', color=color_4)
    else:
        ax.plot(x_point, y_point, marker='o', color=color_5)

    # Triangle
    triangle = Polygon([(x_intercept, 0), (0, y_intercept), (0, 0)], closed=True, edgecolor='#B1B3A0',
                       fill=False,linestyle="--",alpha=0.2)
    ax.add_patch(triangle)

# Set x and y limits based on the custom axis lengths and invert y-axis
ax.set_xlim(0, x_length)
ax.set_ylim(0, y_length)
ax.invert_yaxis()

# Create custom legend entries
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor=color_1, markersize=8, label='<18'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor=color_2, markersize=8, label='18-35'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor=color_3, markersize=8, label='36-53'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor=color_4, markersize=8, label='54-72'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor=color_5, markersize=8, label='72-90')
]

# Add the custom legend
ax.legend(handles=legend_elements, loc='lower right', title='Degree-Color Legend')
ax.get_legend().get_title().set_fontsize('10')


# Add grid lines and show
ax.grid(True)
plt.savefig('dd.png', dpi=300, bbox_inches='tight')
```
