```python
# Load packages and set pi
import math as m
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
pi = m.pi
```

```python
min_x = []
min_y = []
min_d = []
min_a = []
y_values = []
x_values = []

for y in range(1,296):
    print(round(y/296,2))
    for x in range(1,209):
        angle_values = []
        area_values = []
        total_x_values = []
        total_y_values = []
        for angle in np.arange(0.0001, (pi/2)-0.0001, 0.0001):
            y_2 = x/m.tan(angle)
            x_2 = y*m.tan(angle)
            total_x = x_2 + x
            total_y = y_2 + y
            
            # Make sure dimensions fit in an A4 paper 
            
            if total_y<297 and total_x<210:
                area = (total_x*total_y)/2
                area_values.append(area)
                angle_values.append(m.degrees(angle))
                total_x_values.append(total_x)
                total_y_values.append(total_y)
                
            else:
                continue
                
        # Make sure there IS a minimum area to add
        
        if len(area_values)>0:
            min_x.append(total_x_values[area_values.index(min(area_values))])
            min_y.append(total_y_values[area_values.index(min(area_values))])
            min_d.append(angle_values[area_values.index(min(area_values))])
            min_a.append(min(area_values))
            y_values.append(y)
            x_values.append(x)
```

```python
# Background Color
ax = plt.axes()
ax.set_facecolor("#EBE5C1")

# Create the Plot and Grid
plt.scatter(min_d, min_a, color='#031D3A', label='Area')
plt.grid(linestyle='--', alpha = 0.7,c="#E3B441")
plt.ylim(0, 35000)

# Title
plt.suptitle('Distribution of Minimum Areas over Angles', x=0.125, y=0.98, ha='left', fontsize=18)

# Subtitle
plt.title('Points of which a local minimum exists', loc='left', fontsize=14)

# Bottom Text
text = 'The graph above shows the minimum area possible for each possible position of the staple. The angle shows how to attain the minimum area.'
plt.text(0, -6000, text, ha='left', fontsize = 11, 
         alpha=0.9)

# Axis Labels
plt.ylabel('Total Area (mm)',fontsize=15)
plt.xlabel('Angles (degrees)',fontsize=15)

# Annotation Arrow
arrowprops = dict(arrowstyle="->", connectionstyle="angle3,angleA=0,angleB=-90")
plt.annotate('Biggest "Minimum" Area', 
             xy=(min_d[min_a.index(max(min_a))],max(min_a)), 
             xytext=(min_d[min_a.index(max(min_a))]+5, max(min_a)+5),
             arrowprops=arrowprops,
             size = 20)

# Annotation Line
line_y = min_d[min_a.index(max(min_a))]
plt.axvline(line_y, linestyle='--', color='#E3B441', alpha=0.8,linewidth = 2.5)

# Size and Show
plt.rcParams['figure.figsize'] = [18, 6]
plt.xticks(range(0, 90, 5))
plt.savefig('aa.png', dpi=300, bbox_inches='tight')
```

```python
ax = plt.axes()
ax.set_facecolor("#EBE5C1")
plt.hist(min_d, bins=90, edgecolor='#7F9384',color='#074388')
plt.xlabel('Angle Values')
plt.ylabel('Frequency')
plt.title('Histogram of Angle Values')
plt.xticks(range(0, 90, 5))
plt.savefig('bb.png', dpi=300, bbox_inches='tight')
```

```python
## loop

# Create the figure
x_length = 210
y_length = 297
fig, ax = plt.subplots(figsize=(x_length/35, y_length/35))
ax.set_aspect('equal')
ax.set_facecolor("#EBE5C1")

for i in range(0, len(x_values), int(len(x_values)/40)):
    # Coordinates of the point
    x_point = x_values[i]
    y_point = y_values[i]

    x_intercept = min_x[i]
    y_intercept = min_y[i]

    # Plot the point
    ax.plot(x_point, y_point, marker='o', color='#031D3A')

    # Triangle
    triangle = Polygon([(x_intercept, 0), (0, y_intercept), (0, 0)], closed=True, edgecolor='red',
                       fill=False,linestyle="--",alpha=0.3)
    ax.add_patch(triangle)

# Set x and y limits based on the custom axis lengths and invert y-axis
ax.set_xlim(0, x_length)
ax.set_ylim(0, y_length)
ax.invert_yaxis()

# Add grid lines and show
ax.grid(True)
plt.savefig('dd.png', dpi=300, bbox_inches='tight')
```
